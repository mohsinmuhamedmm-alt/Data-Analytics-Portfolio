```python
def get_city(address):
    return address.split(',')[1].strip(' ')
def get_state(address):
    return address.split(',')[2].split(' ')[1]
df['city'] = df['purchase_address'].apply(lambda x: get_city(x) + ' ' + '('+ get_state(x)+')')
df
```

# Hypothesis 3

household repeat purchasing behavior may indicate strong customer retention.

```python
df['purchase_address'].duplicated().sum()
```

```python
household_df = df.groupby('purchase_address').agg({'order_id': 'nunique',
                                                   'revenue': 'sum',
                                                   'quantity_ordered': 'sum',
                                                   'city': 'first'}).reset_index()
household_df
```

```python
household_df.columns=['purchase_address', 'total_orders', 'total_revenue', 'total_quantity', 'city']
household_df
```

```python
household_df['is_repeat'] = household_df['total_orders'] > 1
household_df['avg_order_value'] = household_df['total_revenue'] / household_df['total_orders']
household_df
```

```python
is_repeat = household_df.groupby('is_repeat').agg(households=('purchase_address', 'count'),
                                      avg_orders=('total_orders', 'mean'),
                                      avg_revenue=('total_revenue', 'mean'),
                                      median_revenue=('total_revenue', 'median'),
                                      avg_order_value=('avg_order_value', 'mean'))
is_repeat
```

## Statistical Validation: Repeat vs Non-Repeat Household Revenue

To check whether repeat households generate meaningfully different revenue from one-time households, I compared total household revenue between two groups:

- one-time households
- repeat households

Null hypothesis: there is no difference in average total revenue between the two groups.

Alternative hypothesis: repeat households generate different average total revenue.

```python
repeat_revenue = household_df.loc[household_df['is_repeat'] == True,'total_revenue']

non_repeat_revenue = household_df.loc[ household_df['is_repeat'] == False,'total_revenue']
```

```python
is_repeat.to_csv('repeat.csv', index=False)
```

```python
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(repeat_revenue,non_repeat_revenue,equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print("Result: statistically significant difference")
else:
    print("Result: no statistically significant difference")
```

```python
mean_difference = repeat_revenue.mean() - non_repeat_revenue.mean()

percent_lift = (mean_difference / non_repeat_revenue.mean()) * 100

print("Mean difference:", mean_difference)
print("Percent lift:", percent_lift)
```

```python
household_df['repeat_segment'] = pd.cut(household_df['total_orders'], bins=[0,1,2,999], 
                                         labels=['1 order', '2 orders', '3+ orders' ], include_lowest=True)
household_df[['purchase_address', 'total_orders', 'repeat_segment']].head()
```

```python
repeat_segment_summary= household_df.groupby('repeat_segment', observed= False).agg(household=('purchase_address', 'count'),
                                                                                      avg_orders=('total_orders', 'mean'),
                                                                                      avg_revenue=('total_revenue', 'mean'),
                                                                                      median_revenue=('total_revenue', 'median'),
                                                                                      avg_order_value=('avg_order_value', 'mean'),
                                                                                      total_revenue=('total_revenue', 'sum')).reset_index()
repeat_segment_summary                                                  
```

```python
repeat_segment_summary['revenue_share_pct']=(repeat_segment_summary['total_revenue'] / repeat_segment_summary['total_revenue'].sum()) * 100

repeat_segment_summary['household_share_pct'] =(repeat_segment_summary['household'] / repeat_segment_summary['household'].sum()) *100

repeat_segment_summary
```

```python
repeat_city = household_df.groupby('city').agg(households=('purchase_address','count'),
                                                     repeat_households=('is_repeat','sum'),
                                                     repeat_rate=('is_repeat','mean'),
                                                     total_revenue=('total_revenue','sum')
                                                    )

repeat_city['repeat_rate'] = repeat_city['repeat_rate'] * 100

repeat_city = repeat_city.sort_values('repeat_rate', ascending=False).reset_index()
repeat_city
```

```python
repeat_city.to_csv('repeat_city.csv', index=False)
```

```python
high_repeat_city = household_df.groupby('city').agg(households=('purchase_address', 'count'),
                                                          high_repeat_households_three_orders_plus=('total_orders', lambda x: (x >= 3).sum()),
                                                          high_repeat_rate_three_orders_plus=('total_orders', lambda x: (x >= 3).mean() * 100),
                                                          total_revenue=('total_revenue', 'sum')
                                                         ).sort_values('high_repeat_rate_three_orders_plus', ascending=False).reset_index()

high_repeat_city
```

## Findings

- the housholds that repeated purchases gererates significant higher total revenue around 120%, than the one time purchase households, and the differece is both statistically significant and commercially meaningful. 

- one order households represented 78.1% of all households but generated only 61.9% of total revenue.

- households with 3+ orders represented only 4.1% of households while contributing 10.5% of total revenue.

- average and median household revenue increased across repeat segments, while average order value stayed relatively stable. which suggests that higher cumulative revenue is mainly linked to repeated purchasing frequency rather than bigger individual orders.

- california cities demonstrated the strongest repeat purchasing behavior placing more than one order and high repeat purshasing behavior placing 3 or more order with 65% and 15.5% respectively.

- The business may benefit from focusing more on retaining repeat households, because a smaller group of repeat purchasers contributes disproportionately to revenue.

## Limitation

- purchase address acts only as a household proxy and cannot confirm unique customer identity which may affect the understanding of customer behavior.
- no compaign data, cost per acquisition data or retention cost per client/ household to determine how much from the budget would both acquiring new customer/houshold and retaining them should have.
